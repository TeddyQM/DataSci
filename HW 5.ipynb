Q8:

```python
wordlist = list(words)

six_longest_words = [''] * 6

for word in wordlist:
    for i in range(6):
      if len(word) > len(six_longest_words[i]):
          six_longest_words[i] = word
          break

print("The six longest words in the word list:")
for word in six_longest_words:
    print(word)
```
